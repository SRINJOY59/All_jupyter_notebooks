```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
data = load_iris()
X = data.data[:, [0, 2]] 
def find_clusters(X, eps_range, min_samples_range):
    for eps in eps_range:
        for min_samples in min_samples_range:
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            dbscan.fit(X)
            labels = dbscan.labels_
            core_samples_mask = np.zeros_like(labels, dtype=bool)
            core_samples_mask[dbscan.core_sample_indices_] = True
            n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
            n_noise_ = list(labels).count(-1)
            if n_clusters_ == 3 and n_noise_ <= 10:
                print(eps, min_samples)
                break

if __name__ == "__main__":
    eps_range = np.arange(0.1, 1, 0.01)
    min_samples_range = range(1, 10)
    find_clusters(X, eps_range, min_samples_range)
```

```python
dbscan = DBSCAN(eps=0.30999999999999994, min_samples=4)
dbscan.fit(X)
km = KMeans(n_clusters = 3)
y_means = km.fit_predict(X)
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize = (12,8))
plt.title('Dendrogram Diagram')
dend = shc.dendrogram(shc.linkage(X,method = 'ward'))
plt.show()
```

```python
labels = dbscan.labels_
print(labels)
core_samples_mask = np.zeros_like(labels, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
print(core_samples_mask)
```

```python
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = list(labels).count(-1)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.title(f'DBSCAN Clustering: {n_clusters_} clusters and {n_noise_} noise points')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
plt.scatter(X[y_means == 0,0], X[y_means == 0,1], color = 'red')
plt.scatter(X[y_means == 1,0], X[y_means == 1,1],color = 'green')
plt.scatter(X[y_means == 2,0], X[y_means ==2,1],color = 'blue')
plt.title("Clustering using KMeans Clustering")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 4,affinity = 'euclidean', linkage = 'ward')
```

```python
labels_ = cluster.fit_predict(X)
#plot the cluster
plt.scatter(X[:,0],X[:,1],c = cluster.labels_, cmap = 'rainbow')
plt.xlabel('Petal Length')
plt.ylabel('Sepal Length')
plt.show()
```

```python
df=pd.read_csv('Shimla.csv')
```

```python
df
```

```python
df['Month'] = df['Month'].map({
    'January':1,
    'February':2,
    'March':3,
    'April':4,
    'May':5,
    'June':6,
    'July':7,
    'August':8,
    'September':9,
    'October':10,
    'November':11,
    'December':12
})
```

```python
df['Date'] = pd.to_datetime(df['Year'].astype('str')+'-'+df['Month'].astype('str')+'-'+'01')
df.head()
```

```python
columns_to_convert = [col for col in df.columns if col != 'DateTime' and col != 'Crime_Cases']
```

```python
df = df[columns_to_convert].copy()
```

```python
df['Total_tourist'] = df['Domestic Tourist'] + df['Foreign Tourist']
```

```python
df
```

```python
df.plot()
```

```python
df['Foreign Tourist'].plot()
```

```python
# Replace commas with an empty string and then convert to float
df['Flights (Chandigarh)'] = df['Flights (Chandigarh)'].str.replace(',', '').astype(float)
```

```python
# Replace non-finite values (NaN and inf) with a specific value (e.g., 0)
df['Flights (Chandigarh)'] = df['Flights (Chandigarh)'].replace([np.inf, -np.inf, np.nan], 0)

# Convert the column to integer
df['Flights (Chandigarh)'] = df['Flights (Chandigarh)'].astype(int)
```

```python
df['Flights (Chandigarh)'].dtype
```

```python
df['Year'].dtype
```

```python
from sklearn.impute import KNNImputer
```

```python
from sklearn.impute import KNNImputer

# List of columns to impute (excluding 'DateTime')
columns_to_impute = [col for col in df.columns if col != 'Date']

# Initialize the KNNImputer with the number of neighbors (k)
knn_imputer = KNNImputer(n_neighbors=5)

# Apply the imputer to the selected columns
df[columns_to_impute] = knn_imputer.fit_transform(df[columns_to_impute])
```

```python
df.head(30)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Fit the scaler on your data
scaler.fit(df[['Flights (Chandigarh)', 'Search_Index', 'Domestic Tourist' ,'Foreign Tourist']])

# Transform the data
df[['Flights (Chandigarh)', 'Search_Index', 'Domestic Tourist' ,'Foreign Tourist']] = scaler.transform(df[['Flights (Chandigarh)', 'Search_Index', 'Domestic Tourist' ,'Foreign Tourist']])
```

```python
df['Prev1'] = df['Domestic Tourist'].shift(1)
df['Prev12'] = df['Domestic Tourist'].shift(12)
```

```python
df['Total_tourist'] = df['Domestic Tourist'] + df['Foreign Tourist']
```

```python
X = df[['Flights (Chandigarh)', 'Search_Index','Total_tourist','Foreign Tourist','Prev1','Prev12']]
y = df['Domestic Tourist']
```

```python
X_train = X.iloc[:-24]
y_train = y.iloc[:-24]
X_test = X.iloc[-24:]
y_test = y.iloc[-24:]
```

```python
X_train.tail()
```

```python
from xgboost import XGBRegressor

xgb_regressor = XGBRegressor()
xgb_regressor.fit(X_train, y_train)

# Predict the missing values
y_pred = xgb_regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred))
```

# Naive Bayes Mathematical Understanding ( subtask - 1 )

Naive Bayes methods are a set of supervised learning algorithms based on applying Bayes’ theorem with the “naive” assumption of conditional independence between every pair of features given the value of the class variable. Bayes’ theorem states the following relationship, given class variable and dependent feature vector X1 through Xn :

$$ P(y|X_1,.....,X_n) = (P(y)*P(X_1,......,X_n|y))/P(X_1,......,X_n) $$

Using the naive conditional independence assumption that
$$ P(X_i|y, X_1,....,X_i-1,X_i+1,.....,X_n) = P(X_i|y) $$

for all i, this relationship is simplified to

$$ P(y|X_1,......,X_n) = P(y)*∏ P(X_i|y)/P(X_1,.....X_n)
$$

Since $$P(X_1,.....X_n)$$is constant given the input, we can use the following classification rule:
 
 $$P(y|X_1,......,X_n) ∝ P(y) * P(X_1,......,X_n|y) $$

$$ ŷ = argmax_y P(y) * P(X_1,........,X_n|y) $$

 P(y|X) is the posterior probability of class y given the features X.
 P(X|y) is the likelihood of the features X given class y.
 P(y) is the prior probability of class y.
 P(X) is the probability of the features x, where $$X = {X_1,.......,X_n} $$

From the ŷ value, we can conclude that a query point belongs to which class( In case of binary classification: 0 or 1 ).

## Assumptions 

### Feature Independence

The naive Bayes classifier assumes that the features are conditionally independent given the class label. This means that the presence or absence of one feature does not affect the presence or absence of another feature, given the class label.

### Normal Distribution

 In Gaussian Naive Bayes, it is assumed that the numerical features in each class follow a Gaussian (normal) distribution.

# Gaussian Naive Bayes

In Gaussian Naive Bayes algorithm for classification, The likelihood of the features is assumed to be Gaussian:

$$ P(X_i|y) = \frac{1}{{\sigma \sqrt{2\pi}}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

 μ is the mean of the Gaussian distribution.
 σ is the standard deviation of the Gaussian distribution.
 P(Xi|y) is called the Probability Distribution Function(PDF) for Gaussian Distribution.

Besides Gaussian Naive Bayes, there are other Naive Bayes models exist like Bernoulli Naive Bayes or Multinomial Naive Bayes but watching the present dataset distributions, only Gaussian Naive Bayes is described here.

In summary, the Naive Bayes Classifier is a probabilistic algorithm that makes predictions based on Bayes' theorem, using the assumption of feature independence. Gaussian Naive Bayes is a specific variant of Naive Bayes that assumes numerical features follow a Gaussian distribution. The model learns from the training data by estimating the prior and likelihood probabilities, and it makes predictions by calculating the posterior probability for each class given the input features and selecting the class with the highest probability as the predicted class.

```python
#Standardization
def standardize_data(data):
    mean = np.mean(data, axis=0)
    std = np.std(data, axis=0)
    standardized_data = (data - mean) / std
    return standardized_data
```

# Naive Bayes Class

```python
class NaiveBayes:
    def __init__(self):
        self.priors = {}  # Prior probabilities for each class
        self.means = {}   # Mean of each feature for each class
        self.variances = {}  # Variance of each feature for each class

    def fit(self, X, y):
        n_samples, n_features = X.shape
        unique_classes = np.unique(y)

        for class_name in unique_classes:
            class_mask = y == class_name
            class_samples = X[class_mask]

            self.priors[class_name] = len(class_samples) / n_samples

            self.means[class_name] = np.mean(class_samples, axis=0)
            self.variances[class_name] = np.var(class_samples, axis=0)

    def _calculate_likelihood(self, x, mean, variance):
        exponent = np.exp(-((x - mean) ** 2) / (2 * variance))
        return (1 / np.sqrt(2 * np.pi * variance)) * exponent

    def _calculate_posterior(self, sample):
        posteriors = {}
        for class_name, prior in self.priors.items():
            likelihood = 1.0
            for feature_idx, feature_value in enumerate(sample):
                mean = self.means[class_name][feature_idx]
                variance = self.variances[class_name][feature_idx]
                likelihood *= self._calculate_likelihood(feature_value, mean, variance)
            posterior = prior * likelihood
            posteriors[class_name] = posterior
        return posteriors

    def predict(self, X):
        predictions = []
        for sample in X:
            posteriors = self._calculate_posterior(sample)
            best_class = max(posteriors, key=posteriors.get)
            predictions.append(best_class)
        return predictions
```

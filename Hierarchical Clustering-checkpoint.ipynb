```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
n_students = 100
cgpa = np.random.uniform(6.0, 10.0, size=n_students)
iq = np.random.randint(110, 150, size=n_students)
data = pd.DataFrame({'IQ': iq, 'CGPA': cgpa})

# Print the first few rows of the dataset
print(data.head(100))
import matplotlib.pyplot as plt
plt.scatter(iq,cgpa)
plt.show()
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.figure(figsize = (12,8))
plt.title('CGPA Dendrogram')
dend  = shc.dendrogram(shc.linkage(data,method  = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
labels_ = cluster.fit_predict(data)
km = KMeans(n_clusters = 4)
X = data.iloc[:,:].values
y_means = km.fit_predict(data)

plt.scatter(data['IQ'].values, data['CGPA'].values, c=cluster.labels_, cmap='rainbow')
plt.xlabel('IQ')
plt.ylabel('CGPA')
plt.title('Clustered Data')
plt.show()
```

```python
plt.scatter(X[y_means == 0,0],X[y_means == 0,1],color = 'violet')
plt.scatter(X[y_means == 1,0],X[y_means == 1,1],color = 'green')
plt.scatter(X[y_means == 2,0],X[y_means == 2,1],color = 'red')
plt.scatter(X[y_means == 3,0],X[y_means == 3,1],color = 'yellow')
plt.xlabel('IQ')
plt.ylabel('CGPA')
plt.title('Clustered Data')
plt.show()
```

```python
wcss = []
for i in range(1,15):
    km = KMeans(n_clusters = i)
    km.fit_predict(data)
    wcss.append(km.inertia_)
plt.plot(range(1,15),wcss)
plt.show()
```

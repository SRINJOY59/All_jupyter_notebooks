```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
register_matplotlib_converters()
```

```python
df = pd.read_csv('ice_cream.csv')
```

```python
df
```

```python
df.rename(columns = {'IPN31152N':'Production','DATE':'Date'}, inplace = True)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df
```

```python
df.set_index('Date',inplace = True)
```

```python
df
```

```python
start_date = pd.to_datetime('2000-01-01')
df = df[start_date:]
```

```python
df
```

```python
import matplotlib.pyplot as plt

# Create a figure with a specific size
plt.figure(figsize=(10,4))

# Plot the 'production' column from the DataFrame
plt.plot(df['Production'])

# Add a title and labels for the axes
plt.title('Ice Cream Production over Time', fontsize=20)
plt.ylabel('Production', fontsize=16)

# Add vertical dashed lines for the beginning of each year from 2011 to 2020
for year in range(2001, 2021):
    plt.axvline(pd.to_datetime(str(year) + '-01-01'), color='k', linestyle='--', alpha=0.2)

# Show the plot
plt.show()
```

```python
acf_plot = plot_acf(df['Production'], lags = 48)
```

```python
pacf_plot = plot_pacf(df['Production'])
```

```python
import pandas as pd
data = pd.read_csv('movie_review.csv')
```

```python
import re

def cleantag(text):
    p = re.compile(r'<.*?>')
    return p.sub('', text)
```

```python
data['review'] = data['review'].apply(cleantag)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
import gensim
tokenized_texts = []
for review_text in data['review']:
    tokens = word_tokenize(review_text)
    tokenized_texts.append(tokens)

# Print the tokenized texts
for tokens in tokenized_texts:
    print(tokens)
```

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count = 2,
#     epochs = 100
)
```

```python
model.build_vocab(tokenized_texts)
```

```python
model.train(tokenized_texts, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('wonderful')
```
